```python
# Carga de datos de incidencia delictiva

ruta = '/content/drive/MyDrive/SciData/GEOAES_25/Bases/IDFC_oct2025.csv'

Data = pd.read_csv(ruta, encoding = "cp1252")
Data.head(5)
```

```python
# Añadir columna de total

meses = ['Enero', 'Febrero', 'Marzo',
         'Abril',	'Mayo',	'Junio',
         'Julio',	'Agosto',	'Septiembre',
         'Octubre',	'Noviembre',	'Diciembre']

Data['Total'] = Data[meses].sum(axis=1)

Data.head(5)
```

```python
# Filtrar por delitos de homicidio doloso
Data_filtro = Data.copy()
Data_filtro = Data_filtro[Data_filtro['Subtipo de delito'] == "Homicidio doloso"]
Data_filtro.head(5)
```

```python
# Sumas agrupadas

Data_tot_hom = Data_filtro.copy()
Data_tot_hom = Data_tot_hom.groupby(['Año',	'Clave_Ent',	'Entidad'])['Total'].sum().reset_index()
Data_tot_hom.head(32)
```

```python
# Verificar columnas

Mapa[['CVE_ENT', 'NOMGEO']].dtypes
```

```python
Data_tot_hom[['Clave_Ent', 'Entidad']].dtypes
```

```python
# Combinar datos solo de 2024 con el mapa

Mapa = Mapa.merge(
    Data_tot_hom[Data_tot_hom['Año'] == 2024],
    left_on= ["NOMGEO"],
    right_on=["Entidad"],
    how = 'left'
)
```

```python
Mapa.head(2)
```

```python
# Mapa temático con variable continua

plt.figure(figsize=(15,12), dpi = 500)
Mapa.plot(
    column = 'Total',
    legend = True,
    cmap = 'OrRd',
    legend_kwds = {'label': "Escala Continua",
                   'orientation': 'horizontal'},
    edgecolor = "black"
)
plt.title('Distribución espacial de los homicidios doloso en México\nPor entidad federativa al cierre de 2024')
plt.grid(False)
plt.tight_layout()
ruta_guardado = '/content/drive/MyDrive/SciData/GEOAES_25/Salidas/Mapa1.png'
plt.savefig(ruta_guardado, dpi = 500, bbox_inches = 'tight', facecolor = 'white')
```

```python
# Mapa temático con variable continua

plt.figure(figsize=(15,12), dpi = 500)
Mapa.plot(
    column = 'Total',
    scheme = 'Quantiles',
    k = 5,
    legend = True,
    cmap = 'YlGnBu',
    legend_kwds = {'loc': 'lower left',
                   'fmt': "{:.0f}"},
    edgecolor = "black"
)
plt.title('Distribución espacial de los homicidios doloso en México por quintiles\nPor entidad federativa al cierre de 2024')
plt.grid(False)
plt.tight_layout()
ruta_guardado = '/content/drive/MyDrive/SciData/GEOAES_25/Salidas/Mapa2.png'
plt.savefig(ruta_guardado, dpi = 500, bbox_inches = 'tight', facecolor = 'white')
```
